```python
# Define the StateActionEncoder
class StateActionEncoder(nn.Module):
    def __init__(self, embedding_dim):
        super(StateActionEncoder, self).__init__()
        self.fc1 = nn.Linear(4, 64)  # Increased from 128 to 256
        self.fc2 = nn.Linear(64, 64)  # Added an extra layer
        self.fc3 = nn.Linear(64, embedding_dim)  # Increased from 128 to 512, then to embedding_dim

    def forward(self, state_action):
        x = F.relu(self.fc1(state_action))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Define the StateEncoder
class StateEncoder(nn.Module):
    def __init__(self, embedding_dim):
        super(StateEncoder, self).__init__()
        self.fc1 = nn.Linear(2, 64)  # Increased from 128 to 256
        self.fc2 = nn.Linear(64, 64)  # Added an extra layer
        self.fc3 = nn.Linear(64, embedding_dim)  # Increased from 128 to 512, then to embedding_dim

    def forward(self, state):
        x = F.relu(self.fc1(state))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

class CategoricalEncoder(nn.Module):
    def __init__(self, num_categories, embedding_dim):
        super(CategoricalEncoder, self).__init__()
        self.embedding = nn.Embedding(num_categories, embedding_dim)
        self.fc1 = nn.Linear(embedding_dim, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, embedding_dim)

    def forward(self, categorical_data):
        embedded = self.embedding(categorical_data)
        # print(embedded)
        x = F.relu(self.fc1(embedded))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
def infoNCE_loss(anchor, positive, negatives, temperature, metric_type=1):
    if metric_type == 1:
        # Compute negative L2 distance for the positive pairs
        positive_distances = torch.sum((anchor - positive) ** 2, dim=1, keepdim=True)
        positive_scores = -positive_distances / temperature

        # Compute negative L2 distances for the negative pairs
        negative_distances = torch.sum((anchor.unsqueeze(1) - negatives) ** 2, dim=2)
        negative_scores = -negative_distances / temperature
    else:
        # Use dot product for the positive and negative pairs (default)
        anchor = F.normalize(anchor, dim=-1)
        positive = F.normalize(positive, dim=-1)
        negatives = F.normalize(negatives, dim=-1)

        positive_scores = torch.sum(anchor * positive, dim=1, keepdim=True) / temperature
        negative_scores = torch.matmul(anchor.unsqueeze(1), negatives.transpose(1, 2)).squeeze(1) / temperature

    # print(positive_scores.shape)
    # print(negative_scores.shape)
    # Combine positive and negative scores into logits
    logits = torch.cat([positive_scores, negative_scores], dim=1)  # (batch_size, 1 + num_negatives)

    # Labels are all zeros because the positive examples are always in the first column
    labels = torch.zeros(logits.size(0), dtype=torch.long, device=logits.device)

    # Compute the cross-entropy loss between logits and labels
    return nn.CrossEntropyLoss()(logits, labels)
```

```python
# configs
batch_size = 32
embedding_dim = 16
```

```python
dataset = TrajectoryDataset(maze, 100, embedding_dim=2, num_negatives=10)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
label_dataset = LabelDataset(maze, size=500, embedding_dim=2, num_negatives=10)
label_dataloader = DataLoader(label_dataset, batch_size=batch_size, shuffle=True)
```

```python
encoder_mod1 = StateActionEncoder(embedding_dim)
encoder_mod2 = StateEncoder(embedding_dim)
category_encoder = CategoricalEncoder(label_dataset.num_categories, embedding_dim)

optimizer = optim.Adam(list(encoder_mod1.parameters()) + list(encoder_mod2.parameters()) + list(category_encoder.parameters()), lr=0.001)
```

```python
num_epochs = 32
temperature = 0.1
batch_size = 32
num_negatives = 11
# Initialize encoders
metric_type = 1
eval_type = 'dot'
if metric_type == 1:
    eval_type = 'dist'
```

```python
# Training loop
for epoch in range(num_epochs):
    total_loss = 0

    label_dl_iterator = iter(label_dataloader)
    for anchor, positive, negatives in dataloader:
        try:
            l_anchor, l_positive, l_negatives = next(label_dl_iterator)
        except StopIteration:
            label_dl_iterator = iter(label_dataloader)
            l_anchor, l_positive, l_negatives = next(label_dl_iterator)

        anchor = torch.tensor(anchor).to(torch.float32)
        positive = torch.tensor(positive).to(torch.float32)
        negatives = torch.tensor(negatives).to(torch.float32)
        anchor_enc = encoder_mod1(anchor) # takes state, action tuple
        positive_enc = encoder_mod2(positive) # takes state
        negatives_enc = encoder_mod2(negatives)
        # print(f'negatives shape: {negatives.shape}')
        # print(anchor[:,[2,3]])
        positive_action = anchor[:,[2,3]]
        cur_state = anchor[:,[0,1]]
        angle = torch.arctan2(anchor[:,2], anchor[:,3])
        # print(angle)


        negative_actions = (angle + torch.pi)[:,None] + (torch.rand(num_negatives)[None,:] - 0.5) * (3 * torch.pi / 2)
        negative_dirs = torch.stack([torch.sin(negative_actions), torch.cos(negative_actions)]).moveaxis(0, -1)
        # print(f'negative actions shape: {negative_actions.shape}')
        # print(negative_dirs.shape)
        negative_full = torch.cat((cur_state.unsqueeze(1).expand(-1, num_negatives, -1), negative_dirs), dim=-1)
        # print(negative_full.shape)
        neg_action_enc = encoder_mod1(negative_full)
        # print(f'positive_enc: {positive_enc.shape}, anchor: {anchor_enc.shape}, neg_action_enc: {neg_action_enc.shape}')
        action_loss = infoNCE_loss(positive_enc, anchor_enc, neg_action_enc, temperature, metric_type=metric_type)
        future_loss = infoNCE_loss(anchor_enc, positive_enc, negatives_enc, temperature, metric_type=metric_type)

        l_positive = torch.tensor(l_positive).to(torch.float32)
        l_negatives = torch.tensor(l_negatives).to(torch.float32)
        l_anchor_enc = category_encoder(l_anchor) # takes state, action tuple
        l_positive_enc = encoder_mod2(l_positive) # takes state
        l_negatives_enc = encoder_mod2(l_negatives)
        # print(l_anchor)
        opposite_labels = torch.randint(0, label_dataset.num_categories, size=(l_anchor.shape[0], num_negatives))
        opposite_enc = category_encoder(opposite_labels)
        # print(opposite_enc.shape)
        # print(opposite_labels.shape)

        label_loss = infoNCE_loss(l_anchor_enc, l_positive_enc, l_negatives_enc, temperature, metric_type=metric_type)
        label_loss2 = infoNCE_loss(l_positive_enc, l_anchor_enc, opposite_enc, temperature, metric_type=metric_type)
        # print(f'label_loss: {label_loss}, label_loss2: {label_loss2}'

        loss = label_loss + label_loss2 + action_loss + future_loss
        # print(f'action_loss: {action_loss}, label_loss: {label_loss}')
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    # for anchor, positive, negatives in label_dataloader:
    #   # print(f'label anchor: {anchor}, positive: {positive}, negatives: {negatives}')
    #   positive = torch.tensor(positive).to(torch.float32)
    #   negatives = torch.tensor(negatives).to(torch.float32)
    #   anchor_enc = category_encoder(anchor) # takes state, action tuple
    #   positive_enc = encoder_mod2(positive) # takes state
    #   negatives_enc = encoder_mod2(negatives)

    #   loss = infoNCE_loss(anchor_enc, positive_enc, negatives_enc, temperature, metric_type=0)
    #   optimizer.zero_grad()
    #   loss.backward()
    #   optimizer.step()
    #   total_loss += loss.item()

    print(f'Epoch {epoch+1}, Loss: {total_loss / len(dataloader)}')
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


with torch.no_grad():
    start = torch.tensor([0.5, 0.5])
    language = False
    eval = 'dist'
    vec2 = None
    if language:
        vec2 = category_encoder(torch.tensor(10))
    else:
        goal = torch.tensor([2.5, 3.])
        vec2 = encoder_mod2(goal)
    eps = 30

    env = ContinuousGridEnvironment(maze, start, {})
    fig, ax = plt.subplots()

    def update(i):
        cur_pos = env.agent_position

        activations = []
        angles = torch.linspace(0., 2 * torch.pi, 16)
        for a in angles:
            action = torch.tensor([torch.sin(a), torch.cos(a)])
            vec1 = torch.tensor([cur_pos[0], cur_pos[1], torch.sin(a), torch.cos(a)]).to(torch.float32)
            vec1 = encoder_mod1(vec1)

            if eval == 'dot':
                activations.append((action, F.normalize(vec1, dim=-1) @ F.normalize(vec2, dim=-1).T))
            else:
                activations.append((action, -torch.norm(vec1 - vec2)))

        best_action = activations[np.argmax([x[1].detach().numpy() for x in activations])][0]
        angle = np.arctan2(best_action[0], best_action[1]) + np.random.normal() * eps * (2 * np.pi / 360)
        best_action = torch.tensor(np.array([np.sin(angle), np.cos(angle)]))
        env.move_agent(best_action)

        ax.clear()
        env.display(ax=ax)

    # Set animation display in Jupyter notebook and figure resolution
    plt.rcParams["animation.html"] = "jshtml"
    plt.rcParams['figure.dpi'] = 150

    # Create the animation
    ani = FuncAnimation(fig, update, frames=50, interval=50, repeat=False)

# Display the animation
ani
```
